```python
from langchain_community.document_loaders import JSONLoader
from langchain.embeddings.sentence_transformer import SentenceTransformerEmbeddings
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma
from langchain.docstore.document import Document

#all-MiniLM-L6-v2
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")

documents=[]
idslist = []
i=0
for doc in filtered_list[:100]:
  document = Document(page_content=str(doc), metadata={"source": "local"})
  documents.append(document)
  idslist.append(str(i))
  i+=1

db = Chroma.from_documents(documents, embeddings, collection_metadata={"hnsw:space": "cosine"},ids=idslist)
```

```python
valid_list=(sorted(valid_list, key=lambda x: x['@id']))
```

#CoT

```python
k = 3
example_list = valid_list[278:279] + valid_list[256:257]
for doc in example_list:
  query = "Articles:\n"+ str(doc["t1"]) + "\n\nCase: " + str(doc["t2"])
  print("\033[1m\nQuery:\033[0m ",query)

  docs = db.similarity_search_with_relevance_scores(query,k)
  rele_docs = []
  for j in docs:
      #print((i[0]))
      docinner = eval(j[0].page_content)
      docinstring = "Articles:\n"+ str(docinner["t1"]) + "\n\nCase: " + str(docinner["t2"]) + "\nLabel: " + str(docinner["@label"]) +"\n\n"
      rele_docs.append(docinstring)
      #print((i[0].page_content))

  examples_str = ""
  j=65
  for each in rele_docs:
    examples_str+=f"{chr(j)}. {each}\n"
    j+=1

  template3 = f"""
  You are given a set of examples of <articles, cases, labels> where their corresponding label tells whether the case entails from the articles. Y means entailment. N means not entailed.

  List of {k} examples:
  {examples_str}"""+ """

  Use the examples given above just as a reference.

  You are given a query below consisting of a set of <articles, case>. Y means entailment. N means not entailed. Imagine you are a lawyer, give the intermediate steps, and tell whether the case is entailed from the articles in the format <Label: Y/N>.

  Query: {query}
  """


  prompt3 = PromptTemplate.from_template(template3)

  llm_chain = LLMChain(prompt=prompt3, llm=llm)
  response = llm_chain.invoke(query)['text']

  print("\n\n\033[1mResponse by LLM:\033[0m",response)
  target_output = str(doc["@label"])
  print("\033[1mTarget Label:\033[0m ",target_output)
  print("############################################")
```

```python
example_list = valid_list[258:259]+valid_list[254:255]

k=9
for doc in example_list:
  query = "Articles:\n"+ str(doc["t1"]) + "\n\nCase: " + str(doc["t2"])
  target_output = str(doc["@label"])
  print("\033[1m\nQuery:\033[0m ",query)

  #retrieval
  docs = db.similarity_search_with_relevance_scores(query,k)
  rele_docs = []
  for j in docs:
      #print((i[0]))
      docinner = eval(j[0].page_content)
      docinstring = "Articles:\n"+ str(docinner["t1"]) + "\n\nCase: " + str(docinner["t2"]) + "\nLabel: " + str(docinner["@label"]) +"\n\n"
      rele_docs.append(docinstring)
      #print((i[0].page_content))

  examples_str = ""
  j=65
  for each in rele_docs:
    examples_str+=f"{chr(j)}. {each}\n"
    j+=1
  template3 = f"""
  You are given a set of examples of <articles, cases, labels> where their corresponding label tells whether the case entails from the articles. Y means entailment. N means not entailed.

  List of {k} examples:
  {examples_str}"""+ """

  Use the examples given above just as a reference.

  You are now given a query below consisting of a set of <articles, case>. Imagine you are a lawyer and tell whether the case is entailed from the articles using the following approach IRAL:

  \"
  Issue: define the relevant legal issue from the case in one line
  Rule: apply legal principles from the articles given in the query only and explain the law in two lines,
  Reasoning: develop legal arguments step by step for or against the applicability of the articles to the case and find out whether the case entails from the articles in four sentences.
  Label: Give the final label in the format <Label: Y/N>
  \"

  Use this reasoning approach and tell whether the case is entailed from the articles

  Query: {query}
  """

  prompt3 = PromptTemplate.from_template(template3)
  llm_chain = LLMChain(prompt=prompt3, llm=llm)
  #LLM call
  response = llm_chain.invoke(query)['text']
  print("\n\n\033[1mResponse by LLM:\033[0m",response)



  target_output = str(doc["@label"])
  print("\033[1mTarget Label:\033[0m ",target_output)
  print("############################################")
```

```python
from langchain.memory import ConversationBufferMemory
```

```python
example_list = valid_list[252:253]+valid_list[254:255]
k=9
for doc in example_list:
  query = "Articles:\n"+ str(doc["t1"]) + "\n\nCase: " + str(doc["t2"])
  target_output = str(doc["@label"])
  print("\033[1m\nQuery:\033[0m ",query)

  #retrieval
  docs = db.similarity_search_with_relevance_scores(query,k)
  rele_docs = []
  for j in docs:
      #print((i[0]))
      docinner = eval(j[0].page_content)
      docinstring = "Articles:\n"+ str(docinner["t1"]) + "\n\nCase: " + str(docinner["t2"]) + "\nLabel: " + str(docinner["@label"]) +"\n\n"
      rele_docs.append(docinstring)
      #print((i[0].page_content))

  examples_str = ""
  j=65
  for each in rele_docs:
    examples_str+=f"{chr(j)}. {each}\n"
    j+=1


  template1 = """You are a lawyer performing legal textual entailment task.
  For the below set of articles and case, understand them, and explain them.

  {chat_history}
  Query: {query}
  """

  prompt = PromptTemplate(
      input_variables=["chat_history", "query"], template=template1
  )
  memory = ConversationBufferMemory(memory_key="chat_history")
  llm_chain = LLMChain(prompt=prompt, llm=llm, memory=memory)
  response = llm_chain.invoke(query)['text']
  print("\n\n\033[1mExplanation by LLM:\033[0m",response)




  template2 = """You are a lawyer performing legal textual entailment task.""" + f"""
  You are given a set of examples of <articles, cases, labels> where their corresponding label tells whether the case entails from the articles. Y means entailment. N means not entailed.

  List of {k} examples:
  {examples_str}"""+ """

  Use the examples given above just as a reference.
  Using the explanation solve whether the case is entailed from the articles, for the query given below. Give the answer in the format <Label: Y/N>. Y means entailment, N means not entailed.

  {chat_history}
  Query: {query}
  """

  prompt = PromptTemplate(
      input_variables=["chat_history", "query"], template=template2
  )

  llm_chain = LLMChain(prompt=prompt, llm=llm, memory=memory)
  response = llm_chain.invoke(query)['text']
  print("\n\n\033[1mResponse by LLM:\033[0m",response)


  target_output = str(doc["@label"])
  print("\033[1mTarget Label:\033[0m ",target_output)
  print("\n############################################")
```
